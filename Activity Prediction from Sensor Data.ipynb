# Problem Statement

*This dataset contains the motion data of 14 healthy older aged between 66 and 86 years old, performed broadly scripted activities using a batteryless, wearable sensor on top of their clothing at sternum level. Data is sparse and noisy due to the use of a passive sensor. 
Participants were allocated in two clinical room settings (S1 and S2). The setting of S1 (Room1) uses 4 RFID reader antennas around the room (one on ceiling level, and 3 on wall level) for the collection of data, whereas the room setting S2 (Room2) uses 3 RFID reader antennas (two at ceiling level and one at wall level) for the collection of motion data*

**The activities performed were**: 
* walking to the chair, 
* sitting on the chair, 
* getting off the chair, 
* walking to bed, 
* lying on bed, 
* getting off the bed and 
* walking to the door.

**Hence the possible class labels assigned for every sensor observation are**:
- Sitting on bed
- Sitting on chair
- Lying on bed
- Ambulating, where ambulating includes standing, walking around the room.

**Columns Info**

The content of the file is as follows:
Comma separated values (CSV) format.
* Column 1: Time in seconds 
* Column 2: Acceleration reading in G for frontal axis
* Column 3: Acceleration reading in G for vertical axis
* Column 4: Acceleration reading in G for lateral axis
* Column 5: Id of antenna reading sensor
* Column 6: Received signal strength indicator (RSSI) 
* Column 7: Phase
* Column 8: Frequency
* Column 9: Label of activity, 1: sit on bed, 2: sit on chair, 3: lying, 4: ambulating
* In addition, gender of participant is included in the last character of file name eg: d1p33F (F:female).

```python
# Import the Libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
def data_from_path(path):

    file_list=glob.glob(path+r"\*")
    dfs=[] # Fetches the data link to transfer it to the dataframe
    for i in file_list:
        df=pd.read_csv(i,header=None) # Csv are comma Separated files and so it can read the files.
        dfs.append(df)
    combined_data=pd.concat(dfs,ignore_index=True)
    combined_data.columns=["Time","Acc_f","Acc_v","Acc_l","Antenna_Id","RSSI","Phase","Freq","Label"]
    return combined_data
```

# Predictive Modeling

#### Importing the Models from their respective libraries

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import recall_score,precision_score,accuracy_score,f1_score,classification_report,confusion_matrix

from sklearn.model_selection import train_test_split,GridSearchCV
```

#### Function creation

```python
summary=pd.DataFrame(columns=["Accuracy","Precision","Recall","f1_score"])
summary.index.name="Model Name"
```

```python
def metrics(model,ytest,ypred):
    accuracy=accuracy_score(ytest,ypred)
    precision = precision_score(ytest, ypred, average='weighted')
    recall = recall_score(ytest, ypred, average='weighted')
    f1 = f1_score(ytest, ypred, average='weighted')

    summary.loc[model]=[accuracy,precision,recall,f1]
    return summary
```

# Train Test split

```python
x=final_df.drop(["Label"],axis=1)
y=final_df["Label"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,stratify=y,train_size=0.80,random_state=2)

# Stratify means that the data will be proportionally sorted or distributed in the training and testing data so that it does not cause any issue.
# Train size is 80 % and test size is 20 %
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```
